```r
library(tidyverse)

options(repr.plot.width = 16, repr.plot.height = 12, repr.plot.res = 100)

n <- 200
p <- 0.15
probs <- c()

for (i in 0:n){
  probs <- append(probs, choose(n, i) * p^i * (1 - p)^(n - i))
}

df_factory <- data.frame("defects" = 0:n, "probability" = probs)

ggplot(data = df_factory, aes(x = defects, y = probability)) + 
    geom_bar(stat = "identity", color = "black", fill = "#003f5c", alpha = 0.7) + 
    xlim(0, 50)

1 - sum(df_factory[0:41, "probability"])
1 - pnorm(40, mean = n*p, sd = (n*p*(1-p))^0.5)
1 - pbinom(40, n, p)
```

```r
# HW 5 Problem #1

n <- 25 # sample size
N <- 393 # population total
p <- 0.654 # proportion

N_0 <- p * N #population total with less that 1000 discharges
N_1 <- (1 - p) * N #population total with more than a 1000 discharges

population <- c(rep(0, N_0), rep(1, N_1)) # raw population 

df_data <- matrix(rep(0, N * n), nrow = N, ncol = n) 

set.seed(1)

for (i in 1:N){
    df_data[i, ] <- sample(population, n, replace = TRUE) #sample from the population N times
    }

mu <- 1 - mean(rowMeans(df_data))

ggplot() + geom_density(fill = "#2f4b7c", alpha = 0.4, aes(x = 1 - rowMeans(df_data))) +
           xlab("") + ylab("") + geom_vline(xintercept = mu, color = "red", lwd = 1.5, linetype = "twodash")

print(mu)
```
