```r
data <- as.data.frame(cbind(
x = seq(0, 12, length.out = 500),
chisq_1 = dchisq(x = seq(0, 12, length.out = 500), df = 1),
chisq_3 = dchisq(x = seq(0, 12, length.out = 500), df = 3),
chisq_5 = dchisq(x = seq(0, 12, length.out = 500), df = 5)))

data %>% tail()
```

```r
library(tidyverse)
```

```r
ggplot(data, aes(x, chisq_1)) + geom_line() +
ggtitle(expression({chi^2}[1])) + ylab("pdf") + xlab("x") +
scale_y_continuous(limits = c(0,1)) + scale_x_continuous(limits = c(0, 6)) + 
theme_bw() + theme(plot.title = element_text(hjust = 0.5))
```

```r
ggplot(data, aes(x, chisq_3)) + geom_line() +
ggtitle(expression({chi^2}[3])) + ylab("pdf") + xlab("x") +
scale_y_continuous(limits = c(0,1)) + scale_x_continuous(limits = c(0, 12)) + 
theme_bw() + theme(plot.title = element_text(hjust = 0.5))
```

```r
ggplot(data, aes(x, chisq_5)) + geom_line() +
ggtitle(expression({chi^2}[5])) + ylab("pdf") + xlab("x") +
scale_y_continuous(limits = c(0,1)) + scale_x_continuous(limits = c(0, 12)) + 
theme_bw() + theme(plot.title = element_text(hjust = 0.5))
```

```r
data2 <- as.data.frame(cbind(
                       x = seq(-6,6, length.out = 500),
                       t_1 = dt(x = seq(-6,6, length.out = 500), df = 1),
                       t_3 = dt(x = seq(-6,6, length.out = 500), df = 3),
                       t_10 = dt(x = seq(-6,6, length.out = 500), df = 10)))

t_1 <- ggplot(data2, aes(x, t_1)) + geom_line() + ggtitle(expression(t[1])) +
       ylab("pdf") + xlab("x") + scale_y_continuous(limits = c(0, 0.6)) + 
       theme_bw() + theme(plot.title = element_text(hjust = 0.5))

t_3 <- ggplot(data2, aes(x, t_3)) + geom_line() + ggtitle(expression(t[3])) +
       ylab("pdf") + xlab("x") + scale_y_continuous(limits = c(0, 0.6)) + 
       theme_bw() + theme(plot.title = element_text(hjust = 0.5))

t_10 <- ggplot(data2, aes(x, t_10)) + geom_line() + ggtitle(expression(t[10])) +
       ylab("pdf") + xlab("x") + scale_y_continuous(limits = c(0, 0.6)) + 
       theme_bw() + theme(plot.title = element_text(hjust = 0.5))
```
