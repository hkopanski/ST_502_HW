```r
set.seed(1980)

n <- 500
theta <- 1.2

x <- runif(n, 0, 1)

x_mod <- (theta + 1) * x^theta

ggplot() + geom_density(fill = "#2f4b7c", alpha = 0.4, aes(x = x_mod)) +
           xlab("") + ylab("") + xlim(-0.5, 3) + ylim(0, 0.6)

theta_hats <- rep(0, n)

for (i in 1:n){
    temp_set <- sample(x_mod, 500, replace = TRUE)
    theta_hats[i] <- - 1 - 1 / mean(log(temp_set))
}

ggplot() + geom_histogram(boundary = 0, bins = 50, fill = "#ff7c43", 
                          alpha = 0.4, color = "black", aes(x = theta_hats)) +
           xlab("") + ylab("") + xlim(0, 5) + xlab(expression(widehat(theta))) + ggtitle(theta~ "histogram")

mu <- mean(theta_hats)

se_theta <- sqrt(sum((theta_hats - mu)^2)/n)

print(c(mu, se))

print("The confidence interval is:")
print(c(quantile(theta_hats, 0.025), quantile(theta_hats, 0.975)))
```
