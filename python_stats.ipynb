```python
import math
import seaborn as sns
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

from scipy import stats

n = 200
p = 0.15
k = list(range(0,201))

binom = []

for i in k:
    binom.append(math.comb(n, i)*(p**i)*(1-p)**(n - i))
    
df_binom = pd.DataFrame([k, binom])

df_binom = df_binom.T

print(df_binom)

```

```python
df_binom.columns = ["k", "prob"]
print(df_binom)

# Set the figure size
plt.figure(figsize = (14, 10))

# plot a bar chart
binom_plot = sns.barplot(y = "prob",
                         x = "k",
                         data = df_binom,
                         ci = None, 
                         color = 'maroon');

    
binom_plot.set_xticks(range(0, 201, 10))
binom_plot.set_xlim(0, 60)  
    
print(1 - sum(df_binom["prob"][:40]))

print(stats.norm.cdf(0.2, 0.15, 0.1))
```

```python
sigma_1 = n * 0.15*(1 - 0.15)

print(1 - sum(df_binom["prob"][:40]))
print(1 - stats.binom.cdf(39, 200, 0.15))
print(1 - stats.norm.cdf(39, 30, sigma_1**0.5))
```

```python
import numpy as np 
k = 10
p = 0.5

N = 100

x = np.random.binomial(k, p, size = N)

k_guesses = range(21)

middle_term = (1 - p)**N

for i in range(21):
    x_left = 1 - x / k_guesses[i]
    x_right = 1 - x / (k_guesses[i] + 1)
    print([np.prod(x_left), middle_term, np.prod(x_right)])
```

```python
x_left = 1 - x / k
x_right = 1 - x / (k + 1)
```

```python
print([np.prod(x_left), middle_term, np.prod(x_right)])
```

```python
import math

poss_x = list(range(31))

def f_binom(x, n, p):
    return math.comb(n, x) * p**x * (1 - p)**(n - x)
    
print(map(f_binom, poss_x, 20, 0.6))
```

```python
n = 20
p = 0.6

results = list(map(lambda x: math.comb(n, x) * p**x * (1 - p)**(n - x), poss_x))

print(results)
```
